```python
test_df_drop
```

```python
x = []
y = []
for i in range(10, len(test_df_drop)):
    x.append(test_df_drop.iloc[i-10:i, :6].values.flatten())  # 將前10天的特徵展平
    y.append(test_df_drop.iloc[i, 6])  # 第n天的目標特徵
x, y = np.array(x), np.array(y)
```

```python
print(f"{x.shape=}\t, {type(x)=}\n{y.shape=}\t\t, {type(y)=}")
```

```python
y.shape


y.reshape(-1,1)
```

```python
# 分割數據集為訓練集和測試集
DS_SPLIT=0.8
train_ds_size=round(test_df_drop.shape[0]*DS_SPLIT)

split_index = int(len(x) * DS_SPLIT)
x_train, x_test = x[:split_index], x[split_index:]
y_train, y_test = y[:split_index], y[split_index:]
type(x_train)
```

# Build model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM

model = Sequential()

model.add(LSTM(units=64,return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(LSTM(units=32,return_sequences=False))
model.add(Dense(units=14))
model.add(Dense(units=1))

model.summary()
```

```python
# 編譯模型
model.compile(optimizer='adam', loss='mse')
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# 訓練模型
callback=EarlyStopping( monitor="val_loss",patience=100,restore_best_weights=True)

model.fit(x_train,y_train,
          validation_split=0.1,
          callbacks=[callback],
          epochs=1000,
          batch_size=20
          )
```
